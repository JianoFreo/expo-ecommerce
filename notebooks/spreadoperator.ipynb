````python
//example with objects
const user = {
    name: "Jiano",
    age: 22
};

const copy = {
    ...user
};
````

{

   name: "Jiano",
   
   age: 22
   
}

````python
// example with  arrays
const nums = [1,2,3];

const copy = [...nums];
````

[1,2,3]

````python
// object example
const person = {
    name: "Jiano",
    city: "Manila"
};

console.log({
    ...person,
    age: 22
});
````

{

   name: "Jiano",

   city: "Manila",

   age: 22
   
}



---

# Simple way to think about it

| Syntax          | Meaning                          |
| --------------- | -------------------------------- |
| `{ person }`    | put object inside another object |
| `{ ...person }` | copy properties out of object    |

---

#  Key insight

Spread operator (`...`) does NOT take “the value of the object”.

It takes:

> all **properties inside the object** and spreads them out.

---

# Example with more fields

```js
const person = {
  name: "Jiano",
  city: "Manila"
};
```

### `{ person }`

```js
{
  person: {
    name: "Jiano",
    city: "Manila"
  }
}
```

---

### `{ ...person }`

```js
{
  name: "Jiano",
  city: "Manila"
}
```

---

#  Final one-line idea

* `{ person }` → **wraps**
* `{ ...person }` → **unpacks**
